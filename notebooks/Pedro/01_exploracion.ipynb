### Tasa de suicidios por sexo
* Número de defunciones por 100.000 habitantes
> Debido a que se ha tenido acceso a los datos del Instituto Anatómico Forense de Madrid, se ha introducido una mejora metodológica en el proceso de esta Estadística, que ha permitido asignar de forma más precisa la causa de defunción en las muertes con Intervención Judicial. Como consecuencia, defunciones que estaban asignadas a causas mal definidas han sido reasignadas a causas externas específicas. Por lo tanto, este hecho debe ser tenido en cuenta a la hora de realizar comparaciones con años anteriores a 2013.

```python
df_sui = pd.read_excel('../../data/raw/suicidiosxsexo.xls', header=3)
df_sui = df_sui.rename(columns={'Unnamed: 0': 'com_aut'})
df_sui = df_sui[1:20]
df_sui.com_aut = df_sui.com_aut.str.title().apply(lambda x: x.split(',')[0].split('(')[0]).str.strip()
df_sui
```

*Inicialmente separo los datos en hombres y mujeres, pero posteriormente, prefiero quedarme con los totales.*

```python
df_hom = pd.DataFrame()
df_muj = pd.DataFrame()

for i, col in enumerate(df_sui.columns):
    año = 2000 + i // 3
    if col.startswith('Hom'):
        df_hom_temp = df_sui[['com_aut', col]].rename(columns={col: 'sui_hom'})
        df_hom_temp['año'] = año
        df_hom = pd.concat([df_hom, df_hom_temp])
    elif col.startswith('Muj'):
        df_muj_temp = df_sui[['com_aut', col]].rename(columns={col: 'sui_muj'})
        df_muj_temp['año'] = año
        df_muj = pd.concat([df_muj, df_muj_temp])

pd.merge(df_hom, df_muj, 'left', on=['año', 'com_aut'])
```

```python
df_tot = pd.DataFrame()

for i, col in enumerate(df_sui.columns):
    año = 2000 + i // 3

    if col.startswith('Tot'):
        df_tot_temp = df_sui[['com_aut', col]].rename(columns={col: 'sui'})
        df_tot_temp['año'] = año
        df_tot = pd.concat([df_tot, df_tot_temp])

df_sui = df_tot
```

```python
df_pib = pd.merge(df_pib, df_sui, 'left', ['año', 'com_aut'])
```

*Voy a generar una función para que haga el mismo tratamiento a todos los datos que extraiga de esta página. Inicialmente separaba por sexos, pero finalmente me quedo con los totales.*

```python
def parsear_sanidad(loc:str, nombre:str, sexo:bool, año_inicial:int=2000):

    # === Importar y renombrar columnas y ajustar filas ===
    df = pd.read_excel(loc, header=3)
    df = df.rename(columns={'Unnamed: 0': 'com_aut'})
    df = df[1:20]
    df.com_aut = df.com_aut.str.title().apply(lambda x: x.split(',')[0].split('(')[0]).str.strip()

    # === Covertir al formato que necesito ===
    # === Si solo tiene totales ===
    if not sexo:
        df_ord = pd.DataFrame()
        for i, col in enumerate(df.columns):
            año = año_inicial - 1 + i
            if col.startswith('Tot'):
                df_temp = df[['com_aut', col]].rename(columns={col: nombre})
                df_temp['año'] = año
                df_ord = pd.concat([df_ord, df_temp])
        return df_ord
    
    # === Si tiene sexos ===
    else:
        # df_hom = pd.DataFrame()
        # df_muj = pd.DataFrame()
        df_ord = pd.DataFrame()
        for i, col in enumerate(df.columns):
            año = año_inicial + i // 3
            # if col.startswith('Hom'):
            #     df_hom_temp = df[['com_aut', col]].rename(columns={col: nombre + '_hom'})
            #     df_hom_temp['año'] = año
            #     df_hom = pd.concat([df_hom, df_hom_temp])
            # elif col.startswith('Muj'):
            #     df_muj_temp = df[['com_aut', col]].rename(columns={col: nombre + '_muj'})
            #     df_muj_temp['año'] = año
            #     df_muj = pd.concat([df_muj, df_muj_temp])
            if col.startswith('Tot'):
                df_temp = df[['com_aut', col]].rename(columns={col: nombre})
                df_temp['año'] = año
                df_ord = pd.concat([df_ord, df_temp])

        # return pd.merge(df_hom, df_muj, 'left', on=['año', 'com_aut'])
        return df_ord
```

## Datos de Criminalidad

Datos extraídos de https://www.ine.es/jaxiT3/Tabla.htm?t=74615&L=0 y descargados en csv desde la aplicación web:

### Tasa de criminalidad por 1.000
* Número de Infracciones Penales por 1.000 habitantes

```python
with open('../../data/raw/criminalidadx1000.csv', encoding='utf8') as f:
    criminalidad = []
    for fila in f:
        if fila[:2].isdigit():
            if 'criminalidad' in fila:
                fila_lista = fila.split(';')
                criminalidad.append({'año': int(fila_lista[2]),
                                     'com_aut': fila_lista[0].split(',')[0][3:],
                                     'crim': float(fila_lista[3].replace(',', '.').replace('\n', '')),
                                     })

df_pib = df_pib.merge(pd.DataFrame(criminalidad), 'left', ['año', 'com_aut'])
```
