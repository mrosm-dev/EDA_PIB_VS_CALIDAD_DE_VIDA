| Comunidad Autónoma | Categórica | Identificador territorial de la comunidad autónoma. | Contexto territorial | todas |
| año | Numérica discreta | Año de referencia del indicador. | Dimensión temporal | todas |

| renta_pc | Numérica continua | Renta disponible bruta de los hogares per cápita. Refleja capacidad económica real. | Condiciones económicas (hogares) | renta_mediana, riesgo_pobreza(%), dificultad_fin_mes(%) |
| poblacion | Numérica discreta | Número total de habitantes por comunidad y año. | Contexto demográfico | tasas |

| falta_espacio_vivienda(%) | Numérica continua | Porcentaje de población que vive en viviendas con falta de espacio. | Condiciones materiales | gasto_elevado_vivienda(%), renta_pc |
| retrasos_pagos(%) | Numérica continua | Porcentaje de población con retrasos en pagos de vivienda o suministros. | Seguridad económica | dificultad_fin_mes(%), riesgo_pobreza(%) |
| renta_media | Numérica continua | Renta media por unidad de consumo. Sensible a valores extremos. | Condiciones económicas (ingresos) | renta_mediana, desigualdad_ing(S80/S20) |
| renta_mediana | Numérica continua | Renta mediana por unidad de consumo. Representa el ingreso típico del hogar. | Condiciones económicas (ingresos) | renta_media, riesgo_pobreza(%), dificultad_fin_mes(%) |
| riesgo_pobreza(%) | Numérica continua | Porcentaje de población en riesgo de pobreza (umbral fijo 2008). | Vulnerabilidad económica | renta_mediana, desigualdad_ing(S80/S20) |
| dificultad_fin_mes(%) | Numérica continua | Porcentaje de población con alguna dificultad para llegar a fin de mes. | Seguridad económica | retrasos_pagos(%), inc_gastos_imprevistos(%) |
| desigualdad_ing(S80/S20) | Numérica continua | Cociente entre la renta del 20 % más rico y el 20 % más pobre. | Desigualdad | renta_media vs renta_mediana, riesgo_pobreza(%) |
| inc_gastos_imprevistos(%) | Numérica continua | Porcentaje de población que no puede afrontar gastos imprevistos. | Seguridad económica | dificultad_fin_mes(%), renta_mediana |
| tasa_criminalidad | Numérica continua | Delitos registrados por cada 1.000 habitantes. | Seguridad física y personal | desigualdad_ing(S80/S20), dificultad_fin_mes(%) |
