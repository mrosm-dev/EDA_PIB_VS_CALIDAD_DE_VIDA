```python
import pandas as pd
```

```python
# BIBLIOGRAFIA: https://www.ine.es/jaxiT3/Tabla.htm?t=1448
```

```python
df_esperanza = pd.read_excel("esperanza_vida.xlsx", index_col=0)
```

```python
df_esperanza = df_esperanza.reset_index()
df_esperanza = df_esperanza.rename(columns={df_esperanza.columns[0]: "comunidad"})
```

```python
df_esperanza = df_esperanza.melt(
    id_vars="comunidad",
    var_name="año",
    value_name="esperanza_vida"
)
```

```python
df_esperanza["año"] = pd.to_numeric(df_esperanza["año"], errors="coerce").astype("Int64")
df_esperanza["esperanza_vida"] = pd.to_numeric(df_esperanza["esperanza_vida"], errors="coerce")
df_esperanza["comunidad"] = df_esperanza["comunidad"].astype(str).str.strip()
df_esperanza["comunidad"] = df_esperanza["comunidad"].replace("Andalucia", "Andalucía")
```

```python
#https://www.ine.es/jaxiT3/Tabla.htm?t=1452
df= pd.read_csv("indice_envejecimiento.csv",sep=";")
df = df[df["Comunidades y Ciudades Autónomas"] != "Total Nacional"]
df["Comunidades y Ciudades Autónomas"] = df["Comunidades y Ciudades Autónomas"].str[3:].replace(r",.*$", "", regex=True).str.strip()
df = df.rename(columns={"Comunidades y Ciudades Autónomas":"comunidad","Periodo":"año","Total":"indice_envejecimiento"})
df = df[df["año"]>1999]
```

```python
'''
Índice de envejecimiento: indicador demográfico que expresa la relación entre la población de 65 años o más y la
población menor de 15 años en un territorio. Permite evaluar el grado de envejecimiento de la estructura 
poblacional, donde valores más altos indican una mayor proporción de personas mayores en relación con la población
infantil.
indice de envejecimiento = (Poblacion de 65 años o mas/ Poblacion menor de 15 años ) *100
'''
```

```python
df_envejecimiento = df.copy()
```

```python
df_envejecimiento_esperanza = df_envejecimiento.merge(df_esperanza,on=["comunidad", "año"],how="left")
```

```python
'''
El saldo vegetativo permite evaluar el crecimiento o decrecimiento natural de la población al considerar conjuntamente
nacimientos y defunciones, sin la influencia de los movimientos migratorios. Su análisis resulta especialmente relevante
en contextos de envejecimiento poblacional, ya que refleja de forma directa el impacto combinado de la baja natalidad y
el aumento de la mortalidad, aportando una visión más completa de la dinámica demográfica anual de cada comunidad.
'''
```

```python
# https://www.ine.es/jaxiT3/Tabla.htm?t=1450
df = pd.read_csv("saldo_vegetativo.csv", sep= ";")
df = df[df["Comunidades y Ciudades Autónomas"] != "Total Nacional"]
df["Comunidades y Ciudades Autónomas"] = df["Comunidades y Ciudades Autónomas"].str[3:].replace(r",.*$", "", regex=True).str.strip()
df = df.rename(columns={"Comunidades y Ciudades Autónomas":"comunidad","Periodo":"año","Total":"saldo_vegetativo"})
df = df[df["año"]>1999]
```

```python
df_saldo_vegetativo = df.copy()
```

```python
df_merge = df_envejecimiento_esperanza.merge(df_saldo_vegetativo,on=["comunidad", "año"],how="left")
```

```python
# https://www.ine.es/jaxiT3/Tabla.htm?t=3198&L=0
df = pd.read_csv("edad_media.csv", sep= ";")
df = df[df["Comunidades y Ciudades Autónomas"] != "Total Nacional"]
df["Comunidades y Ciudades Autónomas"] = df["Comunidades y Ciudades Autónomas"].str[3:].replace(r",.*$", "", regex=True).str.strip()
df = df.rename(columns={"Comunidades y Ciudades Autónomas":"comunidad","Periodo":"año","Total":"edad_media"})
df = df[df["año"]>1999]
df = df[df["Sexo"] == "Ambos sexos"]
df = df.drop(columns="Sexo")
```

```python
df_edad_media = df.copy()
```

```python
df_merge2 = df_edad_media.merge(df_merge,on=["comunidad", "año"],how="left")
```

```python
'''
La tasa de crecimiento anual del PIB real per cápita mide cómo varía de un año a otro el nivel medio de producción o renta
por habitante, una vez descontado el efecto de la inflación. Es decir, indica si, en términos reales, cada persona dispone 
en promedio de más o menos riqueza que el año anterior.
'''
```

```python
# https://www.ine.es/jaxiT3/Tabla.htm?t=45599&L=0
df = pd.read_csv("crec_pib_pc_real.csv", sep= ";")
df = df[df["Comunidades y Ciudades Autónomas"] != "Total Nacional"]
df["Comunidades y Ciudades Autónomas"] = df["Comunidades y Ciudades Autónomas"].str.replace(r",.*$", "", regex=True).str.strip()
df = df.rename(columns={"Comunidades y Ciudades Autónomas":"comunidad","Periodo":"año","Total":"tasa_pib_pc"})
df = df[df["año"]>1999]
```

```python
df_tasa_pib_pc = df.copy()
```

```python
df_merge3 = df_merge2.merge(df_tasa_pib_pc,on=["comunidad", "año"],how="left")
```

```python
'''
El PIB por habitante permite comparar el nivel medio de actividad económica entre comunidades autónomas,
eliminando el efecto del tamaño de la población.

Var. PIB (%) es la variación porcentual interanual del PIB, es decir, cuánto crece o cae el PIB respecto 
al año anterior, expresado en porcentaje.
'''
```

```python
# https://datosmacro.expansion.com/pib/espana-comunidades-autonomas?anio=2022
df = pd.read_excel ("pib_varpib.xlsx")
df= df.drop(columns="Unnamed: 3")
df = df.rename(columns={"CCAA":"comunidad","Fecha":"año","PIB anual":"pib_anual(M€)"})
df["comunidad"] = df["comunidad"].str.replace(r"\s*\[\+\]\s*", "", regex=True).str.strip()
df["año"] = df["año"].astype("Int64")
df["pib_anual(M€)"] = df["pib_anual(M€)"].str.replace("M€", "", regex=False)
```

```python
df_merge4 =  df_merge3.merge(df,on=["comunidad", "año"],how="left")
```

```python
# https://datosmacro.expansion.com/pib/espana-comunidades-autonomas?anio=2024
df = pd.read_excel ("pib_pc_varpib.xlsx")
df= df.drop(columns="Unnamed: 3")
df = df.rename(columns={"CCAA":"comunidad","Fecha":"año","PIB Per Capita":"pib_per_capita(€)"})
df["comunidad"] = df["comunidad"].str.replace(r"\s*\[\+\]\s*", "", regex=True).str.strip()
df["año"] = df["año"].astype("Int64")
df["pib_per_capita(€)"] = df["pib_per_capita(€)"].str.replace("€", "", regex=False)
```

```python
df_merge5 = df_merge4.merge(df,on=["comunidad", "año"],how="left")
```
