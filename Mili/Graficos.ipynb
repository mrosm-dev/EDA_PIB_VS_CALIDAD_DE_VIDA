```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/ipc_anual.csv")
```

```python
columnas = df.columns[2:]
comunidades = df["Comunidades y Ciudades Autónomas"].unique()
```

```python
import os

os.makedirs("graficos", exist_ok=True)

for comunidad in comunidades:
    df_and = df[df["Comunidades y Ciudades Autónomas"] == comunidad]
    for columna in columnas:
        plt.plot(
        df_and["año"],
        df_and[columna],
        marker="o")
    
        plt.xticks(df_and["año"], rotation=45)
        plt.xlabel("año")
        plt.ylabel("Índice (base 2021 = 100)")
        plt.title(f"IPC {columna} –{comunidad} (todos los años)")
        plt.grid(True)
        
        nombre = f"graficos/ipc_{columna}_{comunidad}.png"
        plt.savefig(nombre, dpi=150, bbox_inches="tight")
        plt.close()
```

```python
columnas = df.columns[2:]
comunidades_sel = df["Comunidades y Ciudades Autónomas"].unique()

os.makedirs("graficos_juntos", exist_ok=True)

plt.figure(figsize=(10,5))
for columna in columnas:
    for c in comunidades_sel:
        df_c = df[df["Comunidades y Ciudades Autónomas"] == c]
        plt.plot(df_c["año"], df_c[columna], marker="o", label=c)
    
    plt.xlabel("año")
    plt.ylabel("Índice (base 2021 = 100)")
    plt.title(f"IPC {columna} - comparación por comunidades")
    plt.legend()
    plt.grid(True)
    nombre = f"graficos/ipc_{columna}.png"
    plt.savefig(nombre, dpi=150, bbox_inches="tight")
    plt.close()
```

| Qué querés ver       | Cómo interactúan          |
| -------------------- | ------------------------- |
| Evolución conjunta   | Líneas vs año             |
| Comparación puntual  | Scatter (1 año)           |
| Impacto de crisis    | Líneas con cortes         |
| Relación estructural | Correlación por comunidad |
| Desigualdad          | Media vs mediana (líneas) |
